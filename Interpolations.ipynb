### Data manipulation

```python
elec_mixes = xr.open_dataset('elec_mixes.nc')
display(elec_mixes)
```

```python
years_to_add = np.linspace(2020,2050,31)
dates_to_add = np.linspace(1,365,365)
print(years_to_add,
      #dates_to_add
     )
```
